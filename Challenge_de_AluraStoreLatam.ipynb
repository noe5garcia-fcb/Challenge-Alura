### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

print(f'El ingreso total de Tienda 1 es de: ${facturacion_tienda1}')
print(f'El ingreso total de Tienda 2 es de: ${facturacion_tienda2}')
print(f'El ingreso total de Tienda 3 es de: ${facturacion_tienda3}')
print(f'El ingreso total de Tienda 4 es de: ${facturacion_tienda4}')
```

```python
Ingreso_total = facturacion_tienda1 + facturacion_tienda2 + facturacion_tienda3 + facturacion_tienda4
print(f'El ingreso total es de: ${Ingreso_total}')
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]

# gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='blue')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso en $')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
ventas_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_tienda4 = tienda4['Categoría del Producto'].value_counts()

print(f'Las ventas por categoria en la tienda 1 son: {ventas_tienda1}')
print(f'Las ventas por categoria en la tienda 2 son: {ventas_tienda2}')
print(f'Las ventas por categoria en la tienda 3 son: {ventas_tienda3}')
print(f'Las ventas por categoria en la tienda 4 son: {ventas_tienda4}')
```

```python
# Gráfico para Tienda 1
plt.figure(figsize=(10,5))
ventas_tienda1.plot(kind='bar', color='lightcoral')
plt.title('Productos Vendidos por Categoría - Tienda 1')
plt.xlabel('Categoría')
plt.ylabel('Cantidad Vendida')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Diccionario de tiendas
tiendas_dict = {
    1: tienda,
    2: tienda2,
    3: tienda3,
    4: tienda4
}

# Función para mostrar ventas por categoría y graficar
def mostrar_ventas_por_categoria(numero_tienda):
    datos = tiendas_dict.get(numero_tienda)

    if datos is None:
        print(f"❌ Número de tienda {numero_tienda} inválido.")
        return

    # Revisar la columna de categoría
    if 'Categoría del Producto' in datos.columns:
        ventas = datos['Categoría del Producto'].value_counts()
    elif 'Categoría' in datos.columns:
        ventas = datos['Categoría'].value_counts()
    else:
        print(f"❌ Columna de categoría no encontrada en la Tienda {numero_tienda}.")
        return

    print(f"\n📊 Ventas por categoría en la Tienda {numero_tienda}:\n\n{ventas}\n")

    # Gráfica
    plt.figure(figsize=(10, 5))
    ventas.plot(kind='bar', color='mediumseagreen')
    plt.title(f'Productos Vendidos por Categoría - Tienda {numero_tienda}')
    plt.xlabel('Categoría')
    plt.ylabel('Cantidad Vendida')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Mostrar todos los gráficos de las tiendas
def mostrar_todas_las_tiendas():
    for i in range(1, 5):  # Tienda 1, 2, 3, 4
        mostrar_ventas_por_categoria(i)

# Ejecutar la función para mostrar todas las tiendas
mostrar_todas_las_tiendas()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = (tienda['Calificación'].mean())
calificacion_tienda2 = (tienda2['Calificación'].mean())
calificacion_tienda3 = (tienda3['Calificación'].mean())
calificacion_tienda4 = (tienda4['Calificación'].mean())
print(calificacion_tienda1)
print(calificacion_tienda2)
print(calificacion_tienda3)
print(calificacion_tienda4)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Diccionario de tiendas
tiendas_dict = {
    1: tienda,
    2: tienda2,
    3: tienda3,
    4: tienda4
}

# Función para mostrar la calificación promedio de la tienda
def obtener_calificacion_promedio(numero_tienda):
    datos = tiendas_dict.get(numero_tienda)

    if datos is None:
        print(f"❌ Número de tienda {numero_tienda} inválido.")
        return None  # Si la tienda no existe, retornamos None

    # Verificar que la columna 'Calificación' esté presente
    if 'Calificación' in datos.columns:
        promedio = datos['Calificación'].mean()  # Calculamos el promedio
        return promedio
    else:
        print(f"❌ Columna de Calificación no encontrada en la Tienda {numero_tienda}.")
        return None

# Crear listas para los promedios y las tiendas
promedios = []
tiendas = []

# Obtener el promedio de calificación de cada tienda y almacenarlo
for i in range(1, 5):
    promedio = obtener_calificacion_promedio(i)
    if promedio is not None:
        promedios.append(promedio)
        tiendas.append(f'Tienda {i}')

# Crear un DataFrame de pandas para mostrar la tabla
tabla_calificaciones = pd.DataFrame({
    'Tienda': tiendas,
    'Calificación Promedio': promedios
})

# Mostrar la tabla con los promedios de calificación
print("\n📊 Promedio de calificación de cada tienda:")
print(tabla_calificaciones)

# Crear el gráfico con los promedios de calificación de todas las tiendas
plt.figure(figsize=(8, 5))
plt.bar(tiendas, promedios, color='purple')

# Configurar el gráfico
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)  # Ajustamos el eje Y para que el rango esté entre 0 y 5
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# **Introducción**

El presente análisis tiene como objetivo determinar cuál de las cuatro tiendas disponibles representa la mejor opción para que el Sr. Juan pueda vender su producto. Para llegar a esta conclusión, se llevó a cabo un estudio comparativo considerando los siguientes aspectos:

Ingresos totales por tienda

Categorías de productos más y menos vendidas

Calificaciones promedio de los clientes

Producto más y menos vendido

Costo de envío promedio

Este informe resume los hallazgos de cada punto y, con base en ellos, presenta una recomendación final.

# **Desarrollo del análisis**

## Ingresos Totales por Tienda
Se analizaron los ingresos generados por cada tienda, calculando la suma total de ventas (columna “Precio”). Los resultados fueron:

Tienda 1: Alto nivel de facturación

Tienda 2: Facturación media-alta

Tienda 3: Facturación media

Tienda 4: Menor facturación general

Esto indica que la Tienda 1 tiene el mayor volumen de ventas y una buena rotación de productos.

## Ventas por Categoría
Las categorías más vendidas varían por tienda, pero en general:

Tienda 1 y Tienda 2 muestran alta diversidad y volumen en varias categorías.

Tienda 3 y Tienda 4 tienen menor participación o especialización en pocas categorías.

Esto sugiere que la Tienda 1 tiene un público amplio y abierto a diferentes tipos de productos.

## Calificación Promedio
El análisis de la satisfacción del cliente (columna “Calificación”) mostró:

Tienda 2 tiene la calificación promedio más alta.

Tienda 1 y Tienda 3 mantienen un buen nivel.

Tienda 4 obtuvo la menor calificación promedio.

Esto indica que los clientes de la Tienda 2 están más satisfechos con su experiencia de compra.

## Productos Más y Menos Vendidos

Se identificó el producto más vendido: Mesa de noche
Y el menos vendido: Celular ABXY

Este resultado puede estar relacionado con las preferencias del mercado de las tiendas analizadas. La Tienda 1 participó fuertemente en las ventas del producto más popular.

## Costo Promedio de Envío
Se promediaron los costos de envío en cada tienda:

Tienda 3 y Tienda 2 tienen los costos más bajos.

Tienda 1 mantiene un costo medio.

Tienda 4 presenta el envío más costoso.

Esto sugiere que vender en la Tienda 2 o 3 puede beneficiar a los clientes por menor costo logístico.
